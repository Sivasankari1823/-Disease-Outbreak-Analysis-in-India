## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import pandas as pd

df = pd.read_csv("climate health risk data.csv")

df['week_num'] = df['week_of_outbreak'].str.extract(r'(\d+)').astype(int)

df.head()
df.info()
df.describe()
df.isnull().sum()
````

````python
df['Deaths'] = df['Deaths'].fillna(0)
df['preci'] = df['preci'].fillna(df['preci'].median())
df['Temp'] = df['Temp'].interpolate()
df['LAI'] = df['LAI'].interpolate()
df['Cases'] = pd.to_numeric(df['Cases'], errors='coerce')
df['week_of_outbreak'] = df['week_of_outbreak'].astype(str)
df['week_start_date'] = pd.to_datetime(
    df['year'].astype(str) + df['week_num'].astype(str) + '1',
    format='%Y%W%w'
)
df['week_year'] = df['week_start_date'].dt.year
df['week_month'] = df['week_start_date'].dt.month
df['week_day'] = df['week_start_date'].dt.day
df.head()
df.info()
df.describe()
df.isnull().sum()
````

````python
# --- Import libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



def assign_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8, 9]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

df['Season'] = df['mon'].apply(assign_season)

seasonal_cases = df.groupby('Season')['Cases'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=seasonal_cases, x='Season', y='Cases')
plt.title("Seasonal Disease Burden")
plt.xlabel("Season")
plt.ylabel("Total Cases")
plt.show()

````

````python
yearly_cases = df.groupby('year')['Cases'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(yearly_cases['year'], yearly_cases['Cases'], marker='o')
plt.title("Yearly Trend of Total Disease Cases")
plt.xlabel("Year")
plt.ylabel("Total Cases")
plt.grid(True)
plt.show()
````

````python
heatmap_df = df.groupby(['year', 'mon'])['Cases'].sum().reset_index()
heatmap_pivot = heatmap_df.pivot(index='year', columns='mon', values='Cases')

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_pivot, cmap="Reds", linewidths=.5)
plt.title("Heatmap: Cases by Month and Year")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()

````

````python
temp_monthly = df.groupby('mon')['Temp'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(temp_monthly['mon'], temp_monthly['Temp'], marker='o')
plt.title("Average Temperature per Month")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
````

````python
rain_monthly = df.groupby('mon')['preci'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(rain_monthly['mon'], rain_monthly['preci'], marker='o')
plt.title("Average Precipitation per Month")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
````

````python
lai_monthly = df.groupby('mon')['LAI'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(lai_monthly['mon'], lai_monthly['LAI'], marker='o')
plt.title("Average Vegetation Health (LAI) per Month")
plt.xlabel("Month")
plt.ylabel("LAI")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
````

````python
# --- Step 3: Disease-wise EDA ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Top diseases by total cases
disease_sum = df.groupby('Disease', dropna=False)['Cases'].sum().sort_values(ascending=False)
print("Top diseases by total cases:\n", disease_sum.head(10))

plt.figure(figsize=(10,5))
sns.barplot(x=disease_sum.head(10).values, y=disease_sum.head(10).index)
plt.xlabel("Total Cases")
plt.title("Top 10 Diseases by Total Cases")
plt.show()

# 2) Cases distribution (overall) + log view for skewed data
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['Cases'].dropna(), bins=40)
plt.title("Cases distribution")

plt.subplot(1,2,2)
sns.histplot(np.log1p(df['Cases'].dropna()), bins=40)
plt.title("Log(1+Cases) distribution")
plt.show()

# 3) Case severity: Cases vs Deaths scatter
plt.figure(figsize=(7,6))
sns.scatterplot(data=df, x='Cases', y='Deaths', alpha=0.5)
plt.title("Cases vs Deaths (each record)")
plt.xlabel("Cases")
plt.ylabel("Deaths")
plt.show()


````

````python
# --- Step 4: Spatial EDA ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# State-level burden
if 'state_ut' in df.columns:
    state_cases = df.groupby('state_ut')['Cases'].sum().sort_values(ascending=False)
    plt.figure(figsize=(10,5))
    sns.barplot(x=state_cases.head(20).values, y=state_cases.head(20).index)
    plt.xlabel("Total Cases")
    plt.title("Top 20 States/UTs by Cases")
    plt.show()
else:
    print("Column 'state_ut' not found. Adjust name or use 'state'.")

# District hotspots
if 'district' in df.columns:
    district_cases = df.groupby('district')['Cases'].sum().sort_values(ascending=False)
    plt.figure(figsize=(10,5))
    sns.barplot(x=district_cases.head(20).values, y=district_cases.head(20).index)
    plt.xlabel("Total Cases")
    plt.title("Top 20 Districts by Cases")
    plt.show()

# Geo scatter (lat-long). marker size ~ log(cases)
if {'Latitude','Longitude','Cases'}.issubset(df.columns):
    geo = df.groupby(['Latitude','Longitude'], as_index=False)['Cases'].sum()
    plt.figure(figsize=(8,6))
    sizes = np.log1p(geo['Cases'])*8
    plt.scatter(geo['Longitude'], geo['Latitude'], s=sizes, alpha=0.6)
    plt.xlabel("Longitude"); plt.ylabel("Latitude")
    plt.title("Geo scatter: marker size ~ log(total cases at coord)")
    plt.show()
else:
    print("Latitude/Longitude/Cases required for geo scatter not found.")

````

````python
# --- Step 5: Climate EDA ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 1) Joint scatter + trendline for each climate var vs cases
for var in ['Temp','preci','LAI']:
    if var in df.columns:
        plt.figure(figsize=(7,5))
        sns.regplot(x=var, y='Cases', data=df, scatter_kws={'alpha':0.3}, lowess=True)
        plt.title(f"{var} vs Cases (smoothed)")
        plt.show()

# 2) Boxplots of climate variables grouped by season (if Season exists; else by month)
group_col = 'Season' if 'Season' in df.columns else 'mon'
for var in ['Temp','preci','LAI']:
    if var in df.columns and group_col in df.columns:
        plt.figure(figsize=(10,4))
        sns.boxplot(x=group_col, y=var, data=df)
        plt.title(f"{var} distribution by {group_col}")
        plt.show()
````

````python
# --- Step 6: Correlation matrix ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = [c for c in ['Cases','Deaths','Temp','preci','LAI','Latitude','Longitude'] if c in df.columns]
corr = df[num_cols].corr(method='pearson')
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation matrix (numeric vars)")
plt.show()

# Print top correlations with Cases
if 'Cases' in corr.columns:
    corr_with_cases = corr['Cases'].drop('Cases').sort_values(key=lambda x: x.abs(), ascending=False)
    print("Variables most correlated with Cases:\n", corr_with_cases)

````

````python
# --- Step 7: Lag analysis ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create 4 lag versions of rainfall (precipitation)
df['preci_lag1'] = df['preci'].shift(1)
df['preci_lag2'] = df['preci'].shift(2)
df['preci_lag3'] = df['preci'].shift(3)
df['preci_lag4'] = df['preci'].shift(4)

plt.figure(figsize=(12,8))

plt.plot(df['week_num'], df['Cases'], label="Cases", linewidth=2)
plt.plot(df['week_num'], df['preci_lag1'], label="Rainfall Lag 1")
plt.plot(df['week_num'], df['preci_lag2'], label="Rainfall Lag 2")
plt.plot(df['week_num'], df['preci_lag3'], label="Rainfall Lag 3")
plt.plot(df['week_num'], df['preci_lag4'], label="Rainfall Lag 4")

plt.xlabel("Week")
plt.ylabel("Value")
plt.title("Cases vs Rainfall Lag Analysis")
plt.legend()
plt.show()
````

````python
# --- FIX: rename mon -> month ---
df = df.rename(columns={'mon': 'month'})

# --- Create date column ---
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

# --- Step 1: Global threshold ---
global_threshold = df['Cases'].mean() + 2 * df['Cases'].std()

# --- Step 2: Flag outbreaks ---
df['outbreak_global'] = df['Cases'] > global_threshold

# --- Step 3: Count total outbreaks ---
print("Global threshold:", global_threshold)
print("Global outbreak count:", df['outbreak_global'].sum())

# --- Step 4: Show top outbreak rows ---
outbreak_rows = df.loc[df['outbreak_global']].sort_values('Cases', ascending=False)
print(outbreak_rows[['date','state_ut','district','Disease','Cases','Deaths']].head(20))

# --- Step 5: Plot ---
plt.figure(figsize=(10,5))
plt.scatter(
    df['date'], df['Cases'],
    c=df['outbreak_global'].map({True: 'red', False: 'blue'}),
    alpha=0.6
)
plt.title("Outbreak Detection Scatter Plot")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.show()
````
